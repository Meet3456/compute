```python
#3

percent_missing = df.isnull().sum() * 100 / len(df)    #Percentage of null values to the total rows 
percent_missing
```

```python
df2 = df.drop(['Cabin'], axis=1)      # Dropping entire columns having null values more than 50%
df2
```

```python
df3 = df.dropna(axis=0, subset=['Embarked','Age'])            # Dropping null values for columns having less than 50%
df3
```
